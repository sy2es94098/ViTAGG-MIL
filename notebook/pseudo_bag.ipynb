```python
import numpy as np

def get_position_encoding(position, d_model):
    # 計算位置編碼的矩陣
    angle_rads = get_angles(np.arange(position)[:, np.newaxis],
                            np.arange(d_model)[np.newaxis, :],
                            d_model)
    # 使用 sin 和 cos 函數將位置編碼加入特徵向量
    position_encoding = np.zeros(angle_rads.shape)
    position_encoding[:, 0::2] = np.sin(angle_rads[:, 0::2])
    position_encoding[:, 1::2] = np.cos(angle_rads[:, 1::2])
    return position_encoding

def get_angles(position, i, d_model):
    angle_rates = 1 / np.power(10000, (2 * (i // 2)) / np.float32(d_model))
    return position * angle_rates

# 假設有 5 個特徵向量，每個向量維度為 8
feature_vectors = np.random.rand(5, 8)
position_encoding = get_position_encoding(5, 8)

# 將位置編碼加到特徵向量上
encoded_vectors = feature_vectors + position_encoding

# 檢視結果
print("特徵向量：")
print(feature_vectors)
print("\n位置編碼：")
print(position_encoding)
print("\n加上位置編碼後的特徵向量：")
print(encoded_vectors)
```
