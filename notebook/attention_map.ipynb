```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

path = '/data3/ian/dsmil-wsi/dsmil-wsi/05162023/LR_0.0008_BLOSS_0.9_POS_0_HEAD_2_BLOCK_2_LANDMARK_1024_ms_fusion_1_31/postprocess/test_016.png'
img_gray = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

scaler = MinMaxScaler(feature_range=(0, 255))
scaler.fit(img_gray)
img_gray = scaler.transform(img_gray)
img_gray = (np.rint(img_gray)).astype(np.uint8)
print(np.unique(img_gray))

#img_gray = cv2.fromarray(img_gray)
im_color = cv2.applyColorMap(img_gray, cv2.COLORMAP_JET)
im_color = cv2.cvtColor(im_color, cv2.COLOR_BGR2RGB)
plt.imshow(im_color)
plt.show()
```
