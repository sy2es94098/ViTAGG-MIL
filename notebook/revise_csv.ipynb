```python
import csv
import os
path = '/data3/ian/dsmil-wsi/dsmil-wsi/datasets'
sample_path = '/data1/ian/C16_training_small/C16_test_mask/'
src = 'C16_ms_testset_fusion'
des = 'C16_ms_testset_fusion_with_pos'
src_file = os.path.join(path,src,src+'.csv')
des_dir = os.path.join(path,des)
os.mkdir(des_dir)
des_file = os.path.join(des_dir,des+'.csv')

data = []
with open(src_file, 'r') as f:
    csvreader = csv.reader(f)
    for row in csvreader:
        data.append(row)
head = data[0]
data = data[1:]

positive = os.listdir(sample_path)

for i in range(len(data)):
    file_name = os.path.splitext(os.path.basename(data[i][0]))[0] + '.png'
    if file_name in positive:
        data[i][1] = '1'
        
for r in data:
    print(r)

    
with open(des_file, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(head)
    for d in data:
        writer.writerow(d)
```
